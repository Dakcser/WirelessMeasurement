# Show sensor locations

```python
icon_url = 'https://smartcampus.oulu.fi/manage/assets/co2.svg'

m = folium.Map([65.059228, 25.465375], zoom_start=16.45)

sensor_group = folium.FeatureGroup(name="Sensor location").add_to(m)

for cluster in grouped_data:
    for sensor in cluster:
        location = sensor[:2]
        icon = folium.features.CustomIcon(icon_url,icon_size=(14, 14))  # Creating a custom Icon
        sensor_group.add_child(folium.Marker(location=location, icon=icon))  #adding it to the map
        
folium.LayerControl().add_to(m)
```
